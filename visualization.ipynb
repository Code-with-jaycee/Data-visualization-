# MATPLOTLIB Tutorial

All imports

```python
from matplotlib import pylab
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("cinema.csv")
data.head(2)
```

```python
sorted_data = data.sort_values(by=['budget'])
sorted_data.head(2)
```

```python
x = sorted_data["budget"]
y = sorted_data["Gross"]

pylab.plot(x, y,'r') # 'r' stands for red
```

<h4>Drawing a subgraph</h4>
The contents of the brackets represent (rows, columns, indexes)

```python
pylab.subplot(1, 2, 1) 
pylab.plot(x, y, 'r--' )

pylab.subplot(2, 2, 2)
pylab.plot(y, x, 'g*-')


```

# Operator Description

fig.add_axes() = initilializes subplot a = fig.add_subplot(222)<br>

```python
fig = plt.figure()
axis = fig.add_axes([0.5, 0.1, 1, 1]) # Control the left, right, width, height of the canvas (from 0 to 1)
axis.plot(x, y, 'r')
```

Subgraphs

```python
n = np.linspace(0, 10, 25)
m = n * n + 2

fig, axes = plt.subplots(nrows=1, ncols= 2)  # Submap is of 1 row, 2 columns
for ax in axes:
    ax.plot(n, m, 'r')
```

Draring a picture , or graph, inside another graph

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left , right, width, height
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(n, m, 'r-o')
axes2.plot(m, n, 'g')
```

# fig = plt.figure(figsize=(16, 9), dpi=300)

```python
fig = plt.figure(figsize=(16, 9), dpi=300)
fig.add_subplot()
plt.plot(m, n , 'r')
```

```python
fig, axes = plt.subplots()
axes.plot(m, n, 'r')
```

# Adding title and labels

```python
ax.legend(["Label1", "label2"])
fig, axes = plt.subplots()
axes.set_xlabel('x_label')
axes.set_ylabel('y_label')
axes.set_title("title_graph") # Adding title

axes.plot(m, n, 'r')
```
