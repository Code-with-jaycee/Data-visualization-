# MATPLOTLIB Tutorial

All imports

```python
from matplotlib import pylab
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Fifa_world_cup.csv")
data.head(2)
```

```python
data.corr()
```

```python
x = data["total attempts team1"]
y = data["goal preventions team2"]

m = data["conceded team1"]
n = data["goal inside the penalty area team2"]

pylab.plot(x, y,'r') # 'r' stands for red
```

<h4>Drawing a subgraph</h4>
The contents of the brackets represent (rows, columns, indexes)

```python
pylab.subplot(1, 2, 1) 
pylab.plot(x, y, 'r--' )

pylab.subplot(2, 2, 2)
pylab.plot(y, x, 'g*-')


```

# Operator Description

fig.add_axes() = initilializes subplot a = fig.add_subplot(222)<br>
fig, b = plt.subplots(nrows=3, nclos=2) = Adds subplot<br>
ax = plt.subplots(2, 2) = Creates subplot

```python
fig = plt.figure(dpi=400)
axis = fig.add_axes([0.5, 0.1, 1, 1]) # Control the left, right, width, height of the canvas (from 0 to 1)
axis.plot(x, y, 'r')
```

Subgraphs

```python


fig, axes = plt.subplots(nrows=1, ncols= 2, dpi=400)  # Submap is of 1 row, 2 columns
for ax in axes:
    ax.plot(x, y, 'r')
```

Draring a picture , or graph, inside another graph

```python
fig = plt.figure(dpi=400)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left , right, width, height
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.3])

axes1.plot(x, y, 'r-o')
axes2.plot(y, x, 'g')
```

```python
fig = plt.figure(figsize=(16, 9), dpi=300)
fig.add_subplot()
plt.plot(x, y , 'r')
```

```python
fig, axes = plt.subplots(dpi=400)
axes.plot(x, y, 'r')
```

# Adding title and labels

```python
ax.legend(["Label1", "label2"])
fig, axes = plt.subplots(dpi= 200)
axes.set_xlabel('total attempts team1')
axes.set_ylabel('goal preventions team2')
axes.set_title("Fifa World Cup") # Adding title

axes.plot(x,y, 'r')
axes.plot(m,n, 'g-o')
axes.legend(["Attemps vs preventions", "Goal conceed vs Penalty area"], loc=2)
```
