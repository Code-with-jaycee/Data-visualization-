All imports

```python
from matplotlib import pylab
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Fifa_world_cup.csv")
data.head(2)
```

```python
data.corr()
```

```python
x = data["total attempts team1"]
y = data["goal preventions team2"]

m = data["conceded team1"]
n = data["goal inside the penalty area team2"]

pylab.plot(x, y,'r') # 'r' stands for red
```

<h4>Drawing a subgraph</h4>
The contents of the brackets represent (rows, columns, indexes)

```python
pylab.subplot(1, 2, 1) 
pylab.plot(x, y, 'r--' )

pylab.subplot(2, 2, 2)
pylab.plot(y, x, 'g*-')


```

```python


fig, axes = plt.subplots(nrows=1, ncols= 2, dpi=400)  # Submap is of 1 row, 2 columns
for ax in axes:
    ax.plot(x, y, 'r')
```

Draring a picture , or graph, inside another graph

```python
fig = plt.figure(dpi=300)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left , right, width, height
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.3])

axes1.plot(x, y,color='red', lw=0.25 ,marker="o", markerfacecolor='red')
axes1.set_xlabel("total attempts team1")
axes1.set_ylabel("goal preventions team2")
axes1.set_title("Fifa world cup 2022")
axes2.plot(m, n, color='green', lw=0.25 ,marker="o", markerfacecolor='green')
axes2.set_title("Penalty area goals")
```

```python
fig, axes = plt.subplots(dpi=400)
axes.set_title("Fifa world cup")
axes.plot(x, y, color="red", marker="o")
```

```python
ax.legend(["Label1", "label2"])
fig, axes = plt.subplots(dpi= 200)
axes.set_xlabel('total attempts team1')
axes.set_ylabel('goal preventions team2')
axes.set_title("Fifa World Cup") # Adding title

axes.plot(x,y, 'r')
axes.plot(m,n, 'g-o')
axes.legend(["Attemps vs preventions", "Goal conceed vs Penalty area"], loc=2)
```

in matplotlib, you can set other properties such as Line color, transparency and more

```python
fig, axes = plt.subplots(dpi= 400)
axes.plot(x, y, color='red', alpha=.5)
axes.plot(m, n, color='green', alpha=.5)
```

# Line width

```python
fig, axes = plt.subplots(dpi= 400)
axes.plot(x, y, color='red', lw=0.25, linestyle='-')
axes.plot(m, n, color='green', lw = 1, linestyle ='-.' )

line, =axes.plot(x, y, color="black", lw=1.5)
line.set_dashes([5, 10, 4, 10])
```
