Draring a picture , or graph, inside another graph

```python
fig = plt.figure(dpi=300)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left , right, width, height
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.3])

axes1.plot(x, y,color='red', lw=0.25 ,marker="o", markerfacecolor='red')
axes1.set_xlabel("total attempts team1")
axes1.set_ylabel("goal preventions team2")
axes1.set_title("Fifa world cup 2022")
axes2.plot(m, n, color='green', lw=0.25 ,marker="o", markerfacecolor='green')
axes2.set_title("Penalty area goals")
```

# set the canvas grid and axis range

```python
k  = np.linspace(1, 50, 25)
n = np.linspace(1, 20, 25)

fig, axes = plt.subplots(1, 2, dpi=400)

axes[0].plot(k, k**2, k, k**3, lw=2)
axes[0].grid(True)

axes[1].plot(k, k**2, k, k**3, lw=2)
axes[1].grid(True)
axes[1].set_ylim([0, 40000])
axes[1].set_xlim([10, 30])
```

Other 2D Graphics

```python
fig , axes = plt.subplots(1, 4, figsize=(16, 5), dpi=600)
axes[0].set_title("scatter")
axes[0].scatter(k, k**2, color='red')
axes[1].set_title("step plot")
axes[1].step(k, k**3)

axes[2].set_title("bar plot")
axes[2].bar(k, k**4, width=0.8)

axes[3].set_title("fill_between")
axes[3].fill_between(k, k**4, color='green', alpha=0.5)
```

# Drawing countour image
meshgrid (MATLAB Functions) [X,Y] = meshgrid(x,y) transforms the domain specified by vectors x and y into arrays X and Y , which can be used to evaluate functions of two variables and three-dimensional mesh/surface plots.

```python
delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)

X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X-1)**2 - (Y - 1)** 2)
Z = (Z1 - Z2)* 2
```

```python
fig, ax = plt.subplots(dpi=400)
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize= 10)
ax.set_title("Countour map")
```
